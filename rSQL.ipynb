```python
import pandas as pd
import numpy as np
from dateutil import parser
```

```python
dataf= pd.read_csv(r'C:\Users\jyoth\OneDrive\Documents\sql\marketing_campaign_SQL_data.csv')
```

```python
missing_values = dataf.isnull().sum()
missing_values = missing_values[missing_values > 0]
print("\nMissing Values:")
print(missing_values)
print('\nDuplicate Rows:', dataf.duplicated().sum())
print("Dataset Shape:", dataf.shape)
display(dataf.head())
dataf.info()
dataf.describe()
```

```python
dataf.drop_duplicates(inplace=True)
def standardize_date(date_str):
    try:
        parsed_date =  parser.parse(date_str)
        return parsed_date.strftime('%d/%m/%Y')
    except Exception as e:
        return None
dataf['Date'] = dataf['Date'].apply(standardize_date)
dataf['Date'] = pd.to_datetime(dataf['Date'], format='%d/%m/%Y')

dataf['ROI'] = dataf['ROI'].astype(float).round(2)

print("Data cleaning completed successfully!")
dataf.head()
```

```python
dataf.to_csv('marketing_campaign_SQL_data.csv', index=False)
print('saved')
```
